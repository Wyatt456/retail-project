```python
import pandas as pd

# 读取数据
df1 = pd.read_excel("../data/raw/2009-2010.xlsx")
df2 = pd.read_excel("../data/raw/2010-2011.xlsx")

print(df1.shape)
print(df2.shape)
```

```python
df = pd.concat([df1, df2], ignore_index=True) #合并数据 ignore_index=True表示index重新编号

print(df.shape) #查看合并后数据的行列数
```

```python
print(df.info()) #查看数据类型和非空值数量
print(df.describe()) #数值统计
```

```python
missing = df.isnull().sum()

missing_ratio = df.isnull().sum() / len(df) * 100

missing_table = pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_ratio
})

missing_table #检查缺失值
```

```python
df = df.dropna(subset=["Customer ID"]) #删除缺失 Customer ID 的行
```

仔细观察invoice里面的字母可以发现Cancelled order都以C打头

```python
df = df[~df["Invoice"].astype(str).str.startswith("C")] #删除以 "C" 开头的发票
```

```python
df = df[df["Quantity"] > 0] #删除数量小于等于 0 的行（也就是异常值）
```

```python
df = df[df["Price"] > 0] #删除价格小于等于 0 的行（异常值）
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"] #计算每行的总价，新增 TotalPrice 列
```

```python
df.to_csv("../data/cleaned/retail_clean.csv", index=False)
```
